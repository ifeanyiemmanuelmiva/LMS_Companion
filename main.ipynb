```python
import gradio as gr

def create_interface():
    """
    Creates a Gradio interface with a topic text, a video iframe, and a text area.
    """
    with gr.Blocks() as demo:
        gr.Markdown("# Intro to computer science")  # Topic Text
        
        # Video Iframe (simulated with gr.HTML)
        gr.HTML(
            """
            <iframe width="560" height="315" src="https://www.youtube.com/embed/XEjekWehWas?si=RWsneDXaEmBGEUmR" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture; web-share" referrerpolicy="strict-origin-when-cross-origin" allowfullscreen></iframe>
            
            """
        )
        
        gr.Textbox(label="Your Notes", lines=10, placeholder="Enter your thoughts here...") # Text Area

    return demo

if __name__ == "__main__":
    app = create_interface()
    app.launch()
```
